```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

```python
# Custom continuous color palette
continuous_palette = sns.cubehelix_palette(start=0.5, rot=-0.75, gamma=1.0, as_cmap=True)

# Custom one-color sequential palette
one_color_sequential_palette = sns.color_palette(["#f7fbff", "#deebf7", "#c6dbef", "#9ecae1", "#6baed6", "#4292c6", "#2171b5", "#08519c", "#08306b"])

# Custom bipolar sequential palette
bipolar_sequential_palette = sns.color_palette(["#d73027", "#f46d43", "#fdae61", "#fee08b", "#ffffbf", "#d9ef8b", "#a6d96a", "#66bd63", "#1a9850"])
```

```python
def get_spread_palette(base_palette, n_needed):
    """
    Return a list of n_needed colors spread evenly throughout the base_palette list.
    If n_needed >= len(base_palette), fall back to slicing from the beginning.
    """
    palette_size = len(base_palette)
    if n_needed <= 0:
        return []
    if n_needed >= palette_size:
        # If we need more colors than available, just return the full palette
        return base_palette
    
    # Generate indices spaced evenly from 0 to palette_size-1
    indices = np.linspace(3, palette_size - 1, n_needed)
    indices = np.round(indices).astype(int)  # Convert to integer indices
    return [base_palette[i] for i in indices]
```

### **Story 1: Exploring the Trends: Visualizing Titanic Passenger Data**

**Using (Discrete) One Color Sequential palette for this Story**

##### **Task 1**

```python
titanic = sns.load_dataset('titanic')

n_hue_categories = titanic['class'].nunique()  

plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='survived', data=titanic, errorbar=None, palette=get_spread_palette(one_color_sequential_palette, n_hue_categories))
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()

class_counts = titanic['class'].value_counts()
class_counts = class_counts.sort_index()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=get_spread_palette(one_color_sequential_palette, n_hue_categories))
plt.title('Proportion of Passengers in Each Class')
plt.show()
```

```python
survival_counts = titanic.groupby(['class', 'survived'], observed=False).size().unstack()
n_hue_categories = titanic['survived'].nunique()
survival_counts.plot(kind='bar', stacked=True, figsize=(8, 6), color=get_spread_palette(one_color_sequential_palette, n_hue_categories))
plt.title('Survival and Non-Survival Counts by Passenger Class')
plt.ylabel('Number of Passengers')
plt.xlabel('Passenger Class')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

**Stacked Bar Chart:** This additional visualization shows both the survival and non-survival counts within each passenger class. It provides a clear comparison of survival rates within each class, making it easy to see the distribution of survival and non-survival across different classes.

```python
df = pd.read_csv('IPL_Matches_2008_2022.csv')

rivalries = [
    ('Mumbai Indians', 'Chennai Super Kings'),
    ('Royal Challengers Bangalore', 'Chennai Super Kings'),
    ('Royal Challengers Bangalore', 'Mumbai Indians'),
    ('Royal Challengers Bangalore', 'Kolkata Knight Riders'),
    ('Kolkata Knight Riders', 'Mumbai Indians')
]

team_abbr = {
    'Mumbai Indians': 'MI',
    'Chennai Super Kings': 'CSK',
    'Royal Challengers Bangalore': 'RCB',
    'Kolkata Knight Riders': 'KKR'
}

season_parts = {
    '2008-2012': (2008, 2012),
    '2013-2017': (2013, 2017),
    '2018-2022': (2018, 2022)
}

records = {rivalry: {part: {'Team1 Wins': 0, 'Team2 Wins': 0} for part in season_parts} for rivalry in rivalries}

for index, row in df.iterrows():
    team1 = row['Team1']
    team2 = row['Team2']
    winner = row['WinningTeam']
    season = row['Season']
    season = season.split('/')[0]
    
    for rivalry in rivalries:
        if (team1 == rivalry[0] and team2 == rivalry[1]) or (team1 == rivalry[1] and team2 == rivalry[0]):
            for part, (start, end) in season_parts.items():
                if start <= int(season) <= end:
                    if winner == rivalry[0]:
                        records[rivalry][part]['Team1 Wins'] += 1
                    elif winner == rivalry[1]:
                        records[rivalry][part]['Team2 Wins'] += 1

labels = [f'{team_abbr[rivalry[0]]} vs {team_abbr[rivalry[1]]}' for rivalry in rivalries]
x = range(len(rivalries))

fig, ax = plt.subplots(figsize=(15, 10))
bar_width = 0.2
colors = ['b', 'g', 'r']
colors = [bipolar_sequential_palette[0], bipolar_sequential_palette[5], bipolar_sequential_palette[-1]]
offset = 0

for i, part in enumerate(season_parts):
    team1_wins = [records[rivalry][part]['Team1 Wins'] for rivalry in rivalries]
    team2_wins = [records[rivalry][part]['Team2 Wins'] for rivalry in rivalries]
    
    bar1 = ax.bar([p + offset for p in x], team1_wins, bar_width, label=f'Team1 Wins ({part})', color=colors[i])
    bar2 = ax.bar([p + offset for p in x], team2_wins, bar_width, bottom=team1_wins, label=f'Team2 Wins ({part})', color=colors[i], alpha=0.5)
    
    offset += bar_width

ax.set_xlabel('Rivalries (Team1 vs Team2)') 
ax.set_ylabel('Number of Wins')
ax.set_title('Head-to-Head Win/Loss Records of Major Rival Teams Over the Years')
ax.set_xticks([p + bar_width for p in x])
ax.set_xticklabels(labels)
ax.legend()

plt.tight_layout()
plt.show()
```
