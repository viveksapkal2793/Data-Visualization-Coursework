```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
sns.jointplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.suptitle("Jointplot of Iris Dataset with Species Hue", y=1.02)
plt.show()
```

```python
sns.pairplot(data=iris, hue="species", palette="husl")
plt.suptitle("Pairplot of Iris Dataset with Species Hue", y=1.02)
plt.show()
```

```python
dowjones = sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
plt.title("Dow Jones Industrial Average over the years")
plt.show()
```

```python
palette = sns.color_palette("Paired")
life_exp = sns.load_dataset("healthexp").sort_values("Year")
sns.relplot(
    data=life_exp, kind="line",
    x="Spending_USD", y="Life_Expectancy", hue="Country", palette=palette,
    sort=False
)
plt.title("Life Expectancy vs Spending in various couuntries")
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
sns.set_palette("dark")
sns.boxplot(x="cut", y="price", data=diamonds, palette="cubehelix", hue="cut", legend=False)
plt.title("Boxplot of Diamond Price by Cut")
plt.show()
```

```python
sns.countplot(x="color", data=diamonds, palette="Set2", hue="color", legend=False)
plt.title("Count of Diamonds by Color")
plt.show()
```

```python
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen", hue="color", legend=False, palette="muted"
)
plt.title("Diamond Price by Color")
plt.show()
```

```python
sns.displot(data=diamonds, x="carat", kde=True)
plt.title("Distribution of Carats in Diamonds")
plt.show()
```

```python
flights = sns.load_dataset("flights")
pivot_flights = flights.pivot_table(values="passengers", index="month", columns="year", observed=False)
sns.heatmap(data=pivot_flights, cmap="magma")
plt.title("Heatmap of Flight Passengers by Month and Year")
plt.show()
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month", palette="muted", marker="o")
plt.title("Line Plot of Number of Passengers by Year and Month")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.violinplot(x="species", y="flipper_length_mm", data=penguins, palette="Paired", hue="species", legend=False)
plt.title("Violin Plot of Flipper Length by Species (Penguins Dataset)")
plt.show()
```

```python
sns.stripplot(x="species", y="flipper_length_mm", data=penguins, palette="Blues", hue="species", legend=False)
plt.title("Stripplot of Flipper Length by Species (Penguins Dataset)")
plt.show()
```

```python
sns.boxplot(x="species", y="body_mass_g", data=penguins, palette="deep", hue="species", legend=False)
plt.title("Barplot of Body Mass by Species (Penguins Dataset)")
plt.show()
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", palette="Set2", kind="kde", fill=True)
plt.title("Distribution of Flipper Length by Species (Penguins Dataset)")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(8, 6))
sns.countplot(x="survived", hue="pclass", data=titanic, palette="pastel")
plt.title("Survivors vs Non-Survivors by Class in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x="age", hue="survived", kde=True, palette="Blues")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.pairplot(titanic.dropna(), hue="survived", palette="deep", vars=["age", "fare", "pclass"])
plt.suptitle("Pairplot of Titanic Dataset", y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="pclass", hue="sex", palette="husl")
g.map(sns.histplot, "age", kde=True)
g.set_axis_labels("Age", "Survival Count")
g.set_titles("Class {col_name}")
plt.suptitle("Age Distribution by Class and Sex", y=1.02)
plt.show()
```

```python
numeric_titanic = titanic.select_dtypes(include=["float64", "int64"])
corr = numeric_titanic.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, cmap="coolwarm", annot=True)
plt.title("Correlation Heatmap of Titanic Dataset Features")
plt.show()
```
